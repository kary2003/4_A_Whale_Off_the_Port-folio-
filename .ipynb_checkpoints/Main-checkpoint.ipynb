## Data Cleaning
In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.Files:

## Whale Returns
Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_data = Path("Resources/whale.csv")

whale_data = pd.read_csv(
    whale_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
# whale_data.sort_index(inplace=True)
whale_data.head(5).append(whale_data.tail(5))
```

```python
# Count nulls
whale_data.isnull() .sum()
```

```python
# Drop nulls
whale_data.dropna(inplace=True)
whale_data.isnull().sum()
```

 ## Algorithmic Daily Returns
Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_data = Path("Resources/algo.csv")

algo_data = pd.read_csv(algo_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
# algo_data.sort_index(inplace=True)
algo_data.head(5)
```

```python
# Count nulls
algo_data.isnull().sum()
```

```python
# Drop nulls
algo_data.dropna(inplace=True)
algo_data.isnull().sum()
```

## S&P 500 Returns
Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data.

```python
# Reading S&P 500 Closing Prices
sp500_data = Path("Resources/sp500.csv")
sp500_data = pd.read_csv(sp500_data)
sp500_data.head()
```

```python
 # Check Data Types.
sp500_data.dtypes
```

```python
# Fix Data Types
columns = ['Date', 'Close']
sp500_data.columns = columns

sp500_data.set_index(pd.to_datetime(sp500_data['Date'], infer_datetime_format=True), inplace=True)
sp500_data.sort_index(inplace=True)

sp500_data.drop(columns=['Date'], inplace=True)

# Clean identified numeric fields with $ symbol
sp500_data["Close"] = sp500_data["Close"].str.replace("$", "")
sp500_data["Close"]

sp500_data['Close'] = sp500_data['Close'].astype('float')
sp500_data['Close'].dtype
```

```python
 # Calculate the daily return using the 'shift()' function
daily_returns = (sp500_data- sp500_data.shift(1)) / sp500_data.shift(1)
daily_returns.head()
```

```python
 # Rename Column
daily_returns.columns = ['S&P 500']
daily_returns.head()
```

```python
# Drop nulls
daily_returns.dropna(inplace=True)
daily_returns.isnull().sum()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
 # Concatenate all DataFrames into a single DataFrame
joined_data_rows = pd.concat([whale_data, algo_data, daily_returns], axis="rows", join="inner")
joined_data_columns = pd.concat([whale_data, algo_data, daily_returns], axis="columns", join="inner")
joined_data = pd.concat([joined_data_rows, joined_data_columns], axis='columns', join='inner')
joined_data.head()
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)
1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate a rolling window using the exponentially weighted moving average. 
joined_data.ewm(halflife=21).var().plot(figsize=(20,10))
```

```python
 # Calculate annualized Sharpe Ratios
sharpe_ratios = (joined_data.mean()*252) / (joined_data.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot.bar(title="Sharpe Ratios")
```

```python
# Read My_Portfolio_stocks "amzn, pen, iac"
amzn_data = Path("Resources/AMZN.csv")
pen_data = Path("Resources/PEN.csv")
tgt_data = Path("Resources/TGT.csv")
```

```python
 # Read the first stock
amzn_data = pd.read_csv(
    amzn_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
amzn_data.sort_index(inplace=True)
# Rename Column
amzn_data.columns=["AMZN"]
amzn_data.head(5)
```

```python
# read csv
pen_data = pd.read_csv(
    pen_data, index_col="Date", infer_datetime_format=True, parse_dates=True)

pen_data.sort_index(inplace=True)
#Rename Column
pen_data.columns=["PEN"]
pen_data.head(5)
```

```python
# Read csv, index column date, infer datetime format
tgt_data = pd.read_csv(
    tgt_data, index_col="Date", infer_datetime_format=True, parse_dates=True)

tgt_data.sort_index(inplace=True)
# Rename Column
tgt_data.columns=["TGT"]
tgt_data.head()
```

```python
# Concat my_stocks into one DataFrame
my_portfolio = pd.concat([amzn_data, pen_data, tgt_data], axis="columns", join="inner")
my_portfolio.head()
```

```python
# calculate the returns
my_portfolio_returns = (my_portfolio- my_portfolio.shift(1)) / my_portfolio.shift(1)
my_portfolio_returns.head()
```

```python
# count Nulls
my_portfolio_returns.isnull().sum()
```

```python
# Drop Nulls
my_portfolio_returns.dropna(inplace=True)
my_portfolio_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
# YOUR CODE HERE
portfolio_returns = my_portfolio_returns.dot(weights)
portfolio_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
 # Add your "Custom" portfolio to the larger dataframe of fund returns
joined_row = pd.concat([portfolio_returns, whale_data, algo_data, daily_returns], axis="rows", join="inner")
joined_column = pd.concat([portfolio_returns, whale_data, algo_data, daily_returns], axis="columns", join="inner")
data_frame = pd.concat([joined_row, joined_column], axis='columns', join='inner')
data_frame.rename(columns={0:"My_Portfolio"}, inplace=True)
data_frame.head()
```
